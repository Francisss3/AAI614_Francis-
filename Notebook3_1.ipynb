## Random Walk

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re
import time  # For adding a delay between requests

random.seed(datetime.datetime.now().strftime('%s'))

def getLinks(articleUrl):
    html = urlopen(f'http://en.wikipedia.org{articleUrl}')
    bs = BeautifulSoup(html, 'html.parser')
    return bs.find('div', {'id':'bodyContent'}).find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))

links = getLinks('/wiki/Kevin_Bacon')

# Limit the random walk to a maximum number of steps
max_steps = 10
step = 0

while len(links) > 0 and step < max_steps:
    newArticle = links[random.randint(0, len(links)-1)].attrs['href']
    print(newArticle)
    links = getLinks(newArticle)
    step += 1
    time.sleep(1)  # Pause for 1 second to avoid overwhelming the server

print("Random walk completed.")
```

## Recursively crawling an entire site

```python
max_pages = 50  # Limit the total number of pages to crawl
pages = set()

def getLinks(pageUrl, depth=0):
    global pages
    if depth > max_depth or len(pages) >= max_pages:
        return
    try:
        html = urlopen(f'http://en.wikipedia.org{pageUrl}')
        bs = BeautifulSoup(html, 'html.parser')
        for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
            if 'href' in link.attrs:
                if link.attrs['href'] not in pages:
                    newPage = link.attrs['href']
                    print(f"Depth {depth}: {newPage}")
                    pages.add(newPage)
                    time.sleep(1)
                    getLinks(newPage, depth + 1)
    except Exception as e:
        print(f"Error accessing {pageUrl}: {e}")

getLinks('/wiki/Kevin_Bacon')
```

## Collecting Data Across an Entire Site

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

pages = set()
max_pages = 5  # Limit the total number of pages to visit
page_count = 0

def getLinks(pageUrl):
    global page_count
    try:
        html = urlopen(f'http://en.wikipedia.org{pageUrl}')
        bs = BeautifulSoup(html, 'html.parser')

        # Print only the title of the page
        page_title = bs.h1.get_text()
        page_count += 1
        print(f"Page {page_count}: {page_title}")

        # Return links if we haven't hit the max page count
        links = []
        for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
            if 'href' in link.attrs and link.attrs['href'] not in pages:
                links.append(link.attrs['href'])
        return links

    except Exception as e:
        print(f"Error: {e}")
        return []

# Start crawling
queue = ['/wiki/General-purpose_programming_language']  # Start page
while queue and page_count < max_pages:
    current_page = queue.pop(0)  # Get the next page to process
    if current_page not in pages:
        pages.add(current_page)
        new_links = getLinks(current_page)
        queue.extend(new_links[:10])  # Add only the first 10 new links to avoid overload
```

## Crawling across the Internet

```python
from urllib.request import urlopen
from urllib.parse import urlparse
from bs4 import BeautifulSoup
import re
import random
import time

# Retrieves all internal links from a page
def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    scheme = urlparse(url).scheme
    internalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc == '':  # Relative link
            internalLinks.add(f'{scheme}://{netloc}/{link.attrs["href"].strip("/")}')
        elif parsed.netloc == netloc:  # Same domain
            internalLinks.add(link.attrs['href'])
    return list(internalLinks)

# Retrieves all external links from a page
def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    externalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc != '' and parsed.netloc != netloc:  # Different domain
            externalLinks.add(link.attrs['href'])
    return list(externalLinks)

# Gets a random external link
def getRandomExternalLink(startingPage):
    try:
        html = urlopen(startingPage)
        bs = BeautifulSoup(html, 'html.parser')
        externalLinks = getExternalLinks(bs, startingPage)
        if externalLinks:  # If there are external links, return one randomly
            return random.choice(externalLinks)
        else:
            print('No external links found. Searching for internal links...')
            internalLinks = getInternalLinks(bs, startingPage)
            if internalLinks:  # If there are internal links, pick one and retry
                return getRandomExternalLink(random.choice(internalLinks))
            else:
                print('No internal links available either. Stopping.')
                return None
    except Exception as e:
        print(f"Error fetching page: {e}")
        return None

# Follows external links for a limited number of steps
def followExternalOnly(startingSite, max_steps=5):
    step = 0
    currentSite = startingSite
    while step < max_steps:
        print(f"Step {step + 1}: Visiting {currentSite}")
        externalLink = getRandomExternalLink(currentSite)
        if not externalLink:
            print("No more links to follow. Stopping.")
            break
        print(f"Random external link is: {externalLink}")
        currentSite = externalLink
        step += 1
        time.sleep(2)  # Add delay to avoid overloading servers

# Start the crawl
followExternalOnly('https://www.oreilly.com/', max_steps=5)
```

## Collect all External Links from a Site

```python
from urllib.request import urlopen
from urllib.parse import urlparse
from bs4 import BeautifulSoup
import re
import time

# Global lists to track links
allExtLinks = []
allIntLinks = []
max_ext_links = 40  # Stop after collecting 40 external links

# Function to get all internal links from a page
def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    scheme = urlparse(url).scheme
    internalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc == '':  # Relative link
            internalLinks.add(f'{scheme}://{netloc}/{link.attrs["href"].strip("/")}')
        elif parsed.netloc == netloc:  # Same domain
            internalLinks.add(link.attrs['href'])
    return list(internalLinks)

# Function to get all external links from a page
def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    externalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc != '' and parsed.netloc != netloc:  # Different domain
            externalLinks.add(link.attrs['href'])
    return list(externalLinks)

# Recursive function to collect all external links
def getAllExternalLinks(url, max_depth=2, current_depth=0):
    global allExtLinks, allIntLinks, max_ext_links
    if current_depth > max_depth or len(allExtLinks) >= max_ext_links:  # Stop if depth exceeds max_depth or max_ext_links is reached
        return
    try:
        bs = BeautifulSoup(urlopen(url), 'html.parser')
        internalLinks = getInternalLinks(bs, url)
        externalLinks = getExternalLinks(bs, url)

        # Add external links to the list and print them
        for link in externalLinks:
            if len(allExtLinks) >= max_ext_links:  # Stop if the limit is reached
                return
            if link not in allExtLinks:
                allExtLinks.append(link)
                print(f"External Link: {link}")

        # Crawl internal links
        for link in internalLinks:
            if len(allExtLinks) >= max_ext_links:  # Stop if the limit is reached
                return
            if link not in allIntLinks:
                allIntLinks.append(link)
                time.sleep(1)  # Respectful delay
                getAllExternalLinks(link, max_depth, current_depth + 1)

    except Exception as e:
        print(f"Error fetching {url}: {e}")

# Start crawling
start_url = 'https://www.oreilly.com/'
allIntLinks.append(start_url)
getAllExternalLinks(start_url, max_depth=2)  # Limit to 2 levels of depth

print("\nCrawl completed. Total external links collected:", len(allExtLinks))
```
