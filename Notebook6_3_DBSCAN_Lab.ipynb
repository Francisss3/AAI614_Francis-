Plotting Utilities to be used later...

```python
def plot_data(X):
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)

def plot_centroids(centroids, weights=None, circle_color='w', cross_color='k'):
    if weights is not None:
        centroids = centroids[weights > weights.max() / 10]
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='o', s=35, linewidths=8,
                color=circle_color, zorder=10, alpha=0.9)
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='x', s=2, linewidths=12,
                color=cross_color, zorder=11, alpha=1)

def plot_decision_boundaries(clusterer, X, resolution=1000, show_centroids=True,
                             show_xlabels=True, show_ylabels=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                cmap="Pastel2")
    plt.contour(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),
                linewidths=1, colors='k')
    plot_data(X)
    if show_centroids:
        plot_centroids(clusterer.cluster_centers_)

    if show_xlabels:
        plt.xlabel("$x_1$")
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", rotation=0)
    else:
        plt.tick_params(labelleft=False)
```

## DBSCAN

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.05, random_state=42)
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X)
```

```python
dbscan.labels_[:10]
```

```python
dbscan.core_sample_indices_[:10]
```

```python
dbscan.components_
```

```python
def plot_dbscan(dbscan, X, size, show_xlabels=True, show_ylabels=True):
    core_mask = np.zeros_like(dbscan.labels_, dtype=bool)
    core_mask[dbscan.core_sample_indices_] = True
    anomalies_mask = dbscan.labels_ == -1
    non_core_mask = ~(core_mask | anomalies_mask)

    cores = dbscan.components_
    anomalies = X[anomalies_mask]
    non_cores = X[non_core_mask]

    plt.scatter(cores[:, 0], cores[:, 1],
                c=dbscan.labels_[core_mask], marker='o', s=size, cmap="Paired")
    plt.scatter(cores[:, 0], cores[:, 1], marker='*', s=20,
                c=dbscan.labels_[core_mask])
    plt.scatter(anomalies[:, 0], anomalies[:, 1],
                c="r", marker="x", s=100)
    plt.scatter(non_cores[:, 0], non_cores[:, 1],
                c=dbscan.labels_[non_core_mask], marker=".")
    if show_xlabels:
        plt.xlabel("$x_1$")
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", rotation=0)
    else:
        plt.tick_params(labelleft=False)
    plt.title(f"eps={dbscan.eps:.2f}, min_samples={dbscan.min_samples}")
    plt.grid()
    plt.gca().set_axisbelow(True)

dbscan2 = DBSCAN(eps=0.2)
dbscan2.fit(X)

plt.figure(figsize=(9, 3.2))

plt.subplot(121)
plot_dbscan(dbscan, X, size=100)

plt.subplot(122)
plot_dbscan(dbscan2, X, size=600, show_ylabels=False)

plt.show()
```

```python
dbscan = dbscan2  # use eps=0.2
```

```python
from packaging import version
import sklearn
assert version.parse(sklearn.__version__) >= version.parse("1.0.1")

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.cluster import DBSCAN, KMeans
from sklearn.datasets import make_moons
from sklearn.neighbors import KNeighborsClassifier

# Set plot styles
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)
```

```python
def plot_data(X):
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)

def plot_dbscan(dbscan, X, size=100, show_xlabels=True, show_ylabels=True, title=None):
    core_mask = np.zeros_like(dbscan.labels_, dtype=bool)
    if len(dbscan.core_sample_indices_) > 0:  # check to avoid errors if no core points
        core_mask[dbscan.core_sample_indices_] = True
    anomalies_mask = dbscan.labels_ == -1
    non_core_mask = ~(core_mask | anomalies_mask)

    cores = X[core_mask] if len(dbscan.core_sample_indices_) > 0 else np.array([])
    anomalies = X[anomalies_mask]
    non_cores = X[non_core_mask]

    if cores.size > 0:
        plt.scatter(cores[:, 0], cores[:, 1],
                    c=dbscan.labels_[core_mask], marker='o', s=size, cmap="Paired")
        plt.scatter(cores[:, 0], cores[:, 1], marker='*', s=20,
                    c=dbscan.labels_[core_mask])

    if anomalies.size > 0:
        plt.scatter(anomalies[:, 0], anomalies[:, 1],
                    c="r", marker="x", s=100)
    if non_cores.size > 0:
        plt.scatter(non_cores[:, 0], non_cores[:, 1],
                    c=dbscan.labels_[non_core_mask], marker=".")

    if show_xlabels:
        plt.xlabel("$x_1$")
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", rotation=0)
    else:
        plt.tick_params(labelleft=False)

    plt.title(title if title is not None else f"eps={dbscan.eps:.2f}, min_samples={dbscan.min_samples}")
    plt.grid(True)
    plt.gca().set_axisbelow(True)
```

```python
X, y = make_moons(n_samples=1000, noise=0.05, random_state=42)
```

```python
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X)
plt.figure(figsize=(6,4))
plot_dbscan(dbscan, X, size=100, title="DBSCAN (eps=0.05, min_samples=5)")
plt.show()

print("Number of clusters (excluding noise):", len(set(dbscan.labels_)) - (1 if -1 in dbscan.labels_ else 0))
print("Number of noise points:", np.sum(dbscan.labels_ == -1))
```

```python
# try different eps and min_samples values
params = [
    (0.1, 5),
    (0.2, 5),
    (0.2, 10),
    (0.3, 5),
]

fig = plt.figure(figsize=(10, 8))
for i, (eps_val, min_samp) in enumerate(params, start=1):
    db = DBSCAN(eps=eps_val, min_samples=min_samp)
    db.fit(X)
    plt.subplot(2, 2, i)
    plot_dbscan(db, X, size=100, show_xlabels=(i>2), title=f"eps={eps_val}, min_samples={min_samp}")
plt.tight_layout()
plt.show()
```

```python
# Fit K-means with 2 clusters (we know it's a two-moon shape)
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
y_pred_kmeans = kmeans.labels_

# Plot K-means result
plt.figure(figsize=(12,4))

plt.subplot(121)
plot_dbscan(dbscan, X, size=100, title="DBSCAN (eps=0.05, min_samples=5)")
plt.subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=y_pred_kmeans, cmap='Paired', s=30)
plt.title("K-means (k=2)")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.grid(True)
plt.gca().set_axisbelow(True)

plt.tight_layout()
plt.show()

print("DBSCAN cluster labels (unique):", set(dbscan.labels_))
print("K-means cluster labels (unique):", set(y_pred_kmeans))
print("DBSCAN number of clusters (excluding noise):", len(set(dbscan.labels_)) - (1 if -1 in dbscan.labels_ else 0))
print("K-means number of clusters:", 2)
```

```python
# Using the final dbscan model (eps=0.05, min_samples=5) from cell 4
if len(dbscan.core_sample_indices_) > 0:
    core_points = dbscan.components_
    core_labels = dbscan.labels_[dbscan.core_sample_indices_]

    knn = KNeighborsClassifier(n_neighbors=50)
    knn.fit(core_points, core_labels)

    X_new = np.array([[-0.5, 0], [0, 0.5], [1, -0.1], [2, 1]])
    print("KNN Predictions (based on DBSCAN clusters):", knn.predict(X_new))
    print("KNN Probabilities:", knn.predict_proba(X_new))

    # Plot decision boundaries for KNN
    def plot_decision_boundaries_classifier(classifier, X, resolution=1000):
        mins = X.min(axis=0) - 0.1
        maxs = X.max(axis=0) + 0.1
        xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                            np.linspace(mins[1], maxs[1], resolution))
        Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        plt.contourf(xx, yy, Z, cmap="Pastel2")
        plt.contour(xx, yy, Z, colors='k', linewidths=1)
        plt.scatter(X[:,0], X[:,1], c="k", s=10)
        plt.xlabel("$x_1$")
        plt.ylabel("$x_2$", rotation=0)
        plt.grid(True)
        plt.gca().set_axisbelow(True)

    plt.figure(figsize=(6,3))
    plot_decision_boundaries_classifier(knn, X)
    plt.scatter(X_new[:, 0], X_new[:, 1], c="b", marker="+", s=200, zorder=10)
    plt.title("KNN decision boundaries trained on DBSCAN core points")
    plt.show()
else:
    print("No core points identified with the given DBSCAN parameters. Adjust eps/min_samples.")
```
