```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Fetch the webpage
url = "https://www.canada.ca/en/public-health/services/surveillance/respiratory-virus-detections-canada/2018-2019/respiratory-virus-detections-isolations-week-01-ending-january-5-2019.html"
response = requests.get(url)
response.raise_for_status()  # Ensure the request was successful

# Step 2: Parse the webpage content
soup = BeautifulSoup(response.text, 'html.parser')

# Step 3: Locate Table 2 using its id
table = soup.find('table', {'id': 't2'})  # Using the id attribute "t2"

# Step 4: Extract table headers
headers = [header.get_text(strip=True) for header in table.find('thead').find_all('th')]

# Step 5: Extract table rows
rows = []
for row in table.find('tbody').find_all('tr'):  # Focus on tbody rows
    columns = row.find_all('td')
    rows.append([col.get_text(strip=True) for col in columns])

# Step 6: Create a pandas DataFrame
data = pd.DataFrame(rows, columns=headers)

# Step 7: Clean the DataFrame
data = data.dropna(how='all')  # Drop rows with all NaN values
data.columns = data.columns.str.strip()  # Strip extra spaces from headers

# Step 8: Display the DataFrame
print(data)

# Step 9: Save the DataFrame to CSV
data.to_csv("FluWatch_Table2.csv", index=False)
```

```python
from tabulate import tabulate

# Print the DataFrame in a tabular format
print(tabulate(data, headers='keys', tablefmt='grid'))
```
