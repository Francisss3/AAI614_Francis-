```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for numeric variables
students[['Salary', 'Avg_Hours_Studied_per_week', 'GPA']].hist(bins=10, figsize=(15, 5))
plt.suptitle('Histograms of Numeric Variables')
plt.show()

# Count plot for categorical variables
plt.figure(figsize=(12, 5))
sns.countplot(x='University', data=students)
plt.title('Count of Students by University')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(x='Sense of Humour (0-5)', data=students)
plt.title('Count of Students by Sense of Humour')
plt.show()
```

```python
# Scatter plot: Salary vs Avg_Hours_Studied_per_week
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Avg_Hours_Studied_per_week', y='Salary', data=students)
plt.title('Salary vs Avg Hours Studied per Week')
plt.xlabel('Average Hours Studied per Week')
plt.ylabel('Salary')
plt.show()

# Scatter plot: Salary vs GPA
plt.figure(figsize=(8, 6))
sns.scatterplot(x='GPA', y='Salary', data=students)
plt.title('Salary vs GPA')
plt.xlabel('GPA')
plt.ylabel('Salary')
plt.show()
```

```python
# Box plot: Salary by University
plt.figure(figsize=(12, 6))
sns.boxplot(x='University', y='Salary', data=students)
plt.title('Salary by University')
plt.xticks(rotation=45)
plt.show()

# Box plot: Salary by Sense of Humour
plt.figure(figsize=(8, 6))
sns.boxplot(x='Sense of Humour (0-5)', y='Salary', data=students)
plt.title('Salary by Sense of Humour')
plt.show()
```

```python
# Remove dollar sign and 'k', then convert to numeric
students['Salary'] = students['Salary'].replace('[\$,k]', '', regex=True).astype(float) * 1000

# plotting the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(students[['Salary', 'Avg_Hours_Studied_per_week', 'GPA']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
